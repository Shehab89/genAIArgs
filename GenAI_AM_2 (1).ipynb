- The scraped text is not organized and there are random newlines. My goal is to extract the paragraphs from the text as they can be considered the smallest unit for the argument including the claim and permises. To solve this issue, I created a regesx function that remove all the newlines that are not followed by full-stop ".". This approach is not optimal but iit was the best solution so far and I will get back to this point later.

```python

def clean_text(text):
    if not isinstance(text, str):
        return text
    lines = text.split('\n')
    cleaned_lines = []
    for line in lines:
        line = line.rstrip()
        if line.endswith('.'):
            cleaned_lines.append(line + '\n')
        else:
            cleaned_lines.append(line)
    return ''.join(cleaned_lines)

df['body'] = df['body'].apply(clean_text)
```

- Split the text to paragraphs.


```python
def extract_paragraphs(text):
    if not isinstance(text, str):
        return []
    #regex captures sequences of text ended by a full stop and a newline
    paragraphs = re.split(r'(?<=\.)\s*\n', text.strip())
    return paragraphs
```

# Argument Classification:
Identify argumentative paragraphs and their stance towards Generarive AI using pre-trained model by Reimers et al. (2019). This model will classify the paragraphs to "NoArgument", "Argument_against", and "Argument_for".
Paper: https://aclanthology.org/P19-1054/ Code: https://github.com/UKPLab/acl2019-BERT-argument-classification-and-clustering/tree/master/argument-classification

@InProceedings{Reimers:2019:ACL,
  author    = {Reimers, Nils, and Schiller, Benjamin and Beck, Tilman and Daxenberger, Johannes and Stab, Christian and Gurevych, Iryna},
  title     = {{Classification and Clustering of Arguments with Contextualized Word Embeddings}},
  booktitle = {Proceedings of the 57th Annual Meeting of the Association for Computational Linguistics (Volume 1: Long Papers)},
  month     = {07},
  year      = {2019},
  address   = {Florence, Italy},
  pages     = {567--578},
  url       = {https://arxiv.org/abs/1906.09821}
}

```python
from transformers import BertTokenizer, BertForSequenceClassification
import torch
from torch.utils.data import TensorDataset, DataLoader, SequentialSampler
import numpy as np
import pandas as pd

file_path = '/content/filtered_paragraphs_GenAI.csv'
df = pd.read_csv(file_path)

# Model and tokenizer paths
model_path = '/content/drive/MyDrive/ADS Thesis/Model/argument_classification_ukp_all_data'
tokenizer = BertTokenizer.from_pretrained(model_path, do_lower_case=True)

# Define the labels and other parameters
label_list = ["NoArgument", "Argument_against", "Argument_for"]
num_labels = len(label_list)
max_seq_length = 128
eval_batch_size = 8

'''
Create InputExample instances from the dataset
text_a: topic which is here "Generative AI'
text_b: paragraph
'''
class InputExample(object):
    def __init__(self, text_a, text_b=None, label=None):
        self.text_a = text_a
        self.text_b = text_b
        self.label = label

def _truncate_seq_pair(tokens_a, tokens_b, max_length):
    """Truncates a sequence pair in place to the maximum length."""
    while True:
        total_length = len(tokens_a) + len(tokens_b)
        if total_length <= max_length:
            break
        if len(tokens_a) > len(tokens_b):
            tokens_a.pop()
        else:
            tokens_b.pop()

'''
BERT expects input data in a structured format, which includes:

Input IDs: These are numerical representations of tokens (words or subwords) in the text. Each token is mapped to a unique ID in the tokenizer's vocabulary.
Attention Masks: These indicate which tokens should be attended to (i.e., considered) by the model. A value of 1 means the token is part of the input sequence, and 0 means it's padding.
Segment IDs: These differentiate between the two sentences in a pair when processing sentence pairs. BERT uses these IDs to understand which tokens belong to the first sentence and which belong to the second.
'''
def convert_examples_to_features(examples, label_list, max_seq_length, tokenizer):
    label_map = {label: i for i, label in enumerate(label_list)}
    features = []
    for (ex_index, example) in enumerate(examples):
        tokens_a = tokenizer.tokenize(example.text_a)
        tokens_b = None
        if example.text_b:
            tokens_b = tokenizer.tokenize(example.text_b)
            _truncate_seq_pair(tokens_a, tokens_b, max_seq_length - 3)
        else:
            if len(tokens_a) > max_seq_length - 2:
                tokens_a = tokens_a[:(max_seq_length - 2)]

        tokens = ["[CLS]"] + tokens_a + ["[SEP]"] #CLS indicates the begining of the sentence and SEP indicates either the end of the sentence or seprates two senteences.
        segment_ids = [0] * len(tokens)

        if tokens_b:
            tokens += tokens_b + ["[SEP]"]
            segment_ids += [1] * (len(tokens_b) + 1)

        input_ids = tokenizer.convert_tokens_to_ids(tokens)
        input_mask = [1] * len(input_ids)

        padding = [0] * (max_seq_length - len(input_ids))
        input_ids += padding
        input_mask += padding
        segment_ids += padding

        features.append({
            'input_ids': input_ids,
            'input_mask': input_mask,
            'segment_ids': segment_ids,
            'label_id': label_map[example.label] if example.label else None
        })
    return features

#Convert paragraphs to input examples with defining the topic as "Generative AI"
examples = [InputExample(text_a='Generative AI', text_b=paragraph) for paragraph in df['paragraphs']]
features = convert_examples_to_features(examples, label_list, max_seq_length, tokenizer)

#Create DataLoader for evaluation
all_input_ids = torch.tensor([f['input_ids'] for f in features], dtype=torch.long)
all_input_mask = torch.tensor([f['input_mask'] for f in features], dtype=torch.long)
all_segment_ids = torch.tensor([f['segment_ids'] for f in features], dtype=torch.long)

#Creat TensorDataset
eval_data = TensorDataset(all_input_ids, all_input_mask, all_segment_ids)
eval_sampler = SequentialSampler(eval_data)
eval_dataloader = DataLoader(eval_data, sampler=eval_sampler, batch_size=eval_batch_size)

#Load the pre-trained model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = BertForSequenceClassification.from_pretrained(model_path, num_labels=num_labels)
model.to(device)
model.eval()

#Classify the paragraphs
predicted_labels = []
with torch.no_grad():
    for input_ids, input_mask, segment_ids in eval_dataloader:
        input_ids = input_ids.to(device)
        input_mask = input_mask.to(device)
        segment_ids = segment_ids.to(device)

        logits = model(input_ids, attention_mask=input_mask, token_type_ids=segment_ids).logits
        logits = logits.detach().cpu().numpy()

        for prediction in np.argmax(logits, axis=1):
            predicted_labels.append(label_list[prediction])

df['predicted_label'] = predicted_labels
```
